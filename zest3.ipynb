```python
def draw(x_, y_, name1_, name_, xtitle_, ytitle_, xe_ = [], ye_ = []):
    trace = go.Scatter(
        x = x_,
        y = y_,
        mode = 'lines+markers',
        name = name1_
    )
    
    exact_trace = go.Scatter(
        x = xe_,
        y = ye_,
        mode = 'lines+markers',
        name = "Dokładny"
    )
    
    layout_ = go.Layout(
        title = name_,
            xaxis = dict(
                title = xtitle_,
                titlefont = dict(
                    family = 'Courier New, monospace',
                    size = 18,
                    color = '#7f7f7f'
                )
            ),
            yaxis=dict(
                title = ytitle_,
                titlefont = dict(
                    family = 'Courier New, monospace',
                    size = 18,
                    color = '#7f7f7f'
                )
            )
        )

    data_ = []
    if xe_ != [] and ye_ !=[]:
        data_ = [trace, exact_trace]
    else:
        data_ = [trace]
        
    fig = go.Figure(data = data_, layout = layout_)
    url = py.plot(fig, filename = name_ + name1_)
    print(url)
```

```python
def get_x(m, k):
    size = 100
    xb = [(x*m)/(2*k) for x in range(0, math.ceil((size*2*k)/m))]         # m/(2*k)
    xc = [(x*m)/k for x in range(0, math.ceil((size*k)/m))]               # m/k
    xd = [(x*3*m)/(2*k) for x in range(0, math.ceil((size*2*k)/(3*m)))]   # (3*m)/(2*k)
    xe = [x*2*k*m for x in range(0, math.ceil(size/(2*k*m)))]             # (2*m)/k
    xf = [x*3*k*m for x in range(0, math.ceil(size/(3*k*m)))]             # (3*m)/k
    
    return xb, xc, xd, xe, xf
```

```python
def simulate_zad1_exact(dt, l):
    k = 1
    m = 10
    tt = [dt*i for i in range(0, l)]
    vt = [10]
    
    for i in range(1, l):
        vt.append(vt[0]*math.exp((-k*tt[i])/m))
    
    return tt, vt
```

```python
def simulate_zad2_Euler(dt):
    k = 1      #spring const
    m = 1      #mass
    xtmp = 1   #postion
    vtmp = 0   #velocity
    t = dt      #time
    stopT = 30 #max Time
    
    vt = [0]
    xt = [1]
    tt = [0]
    VP = [1/2*k*xt[len(xt)-1]*xt[len(xt)-1]]
    VK = [0.5*m*vt[len(vt)-1]*vt[len(vt)-1]]
    V = [VP[len(VP)-1]+VK[len(VK)-1]]
    
    while(t <= stopT):
        vt.append(vtmp + dt*(-k/m)*xtmp)
        xt.append(xtmp + dt*vtmp)
        tt.append(t)
        VP.append(1/2*k*xt[len(xt)-1]*xt[len(xt)-1])
        VK.append(0.5*m*vt[len(vt)-1]*vt[len(vt)-1])
        V.append(VP[len(VP)-1]+VK[len(VK)-1])
        xtmp = xt[len(xt)-1]
        vtmp = vt[len(vt)-1]
        t += dt
            
    return tt, xt, vt, VP, VK, V
```

```python
def simulate_zad2_modified_Euler(dt):
    k = 1      #spring const
    m = 1      #mass
    xtmp = 1   #postion
    vtmp = 0   #velocity
    t = dt      #time
    stopT = 30 #max Time
    
    vt = [0]
    xt = [1]
    tt = [0]
    VP = [1/2*k*xt[len(xt)-1]*xt[len(xt)-1]]
    VK = [0.5*m*vt[len(vt)-1]*vt[len(vt)-1]]
    V = [VP[len(VP)-1]+VK[len(VK)-1]]
    
    while(t <= stopT):
        vt.append(vtmp + dt*(-k/m)*xtmp)
        xt.append(xtmp + dt*vt[len(vt)-1])
        tt.append(t)
        VP.append(1/2*k*xt[len(xt)-1]*xt[len(xt)-1])
        VK.append(0.5*m*vt[len(vt)-1]*vt[len(vt)-1])
        V.append(VP[len(VP)-1]+VK[len(VK)-1])
        xtmp = xt[len(xt)-1]
        vtmp = vt[len(vt)-1]
        t += dt
            
    return tt, xt, vt, VP, VK, V
```

```python
def simulate_zad2_Verlet(dt):
    k = 1      #spring const
    m = 1      #mass
    xtmp = 1   #postion
    xtmp2 = 1  #postion
    vtmp = 0   #velocity
    t = dt      #time
    stopT = 30 #max Time
    
    vt = [0]
    xt = [1]
    tt = [0]
    VP = [1/2*k*xt[len(xt)-1]*xt[len(xt)-1]]
    VK = [0.5*m*vt[len(vt)-1]*vt[len(vt)-1]]
    V = [VP[len(VP)-1]+VK[len(VK)-1]]
    
    while(t <= stopT):  
        vt.append((xtmp-xtmp2)/dt)
        xt.append(-xtmp2 +2*xtmp - dt*dt*(k/m)*xtmp)
        tt.append(t)
        VP.append(1/2*k*xt[len(xt)-1]*xt[len(xt)-1])
        VK.append(0.5*m*vt[len(vt)-1]*vt[len(vt)-1])
        V.append(VP[len(VP)-1]+VK[len(VK)-1])
        xtmp2 = xtmp
        xtmp = xt[len(xt)-1]
        vtmp = vt[len(vt)-1]
        t += dt
            
    return tt, xt, vt, VP, VK, V
```

```python
def simulate_zad2_exact(dt):
    t = dt
    fi = 0
    m = 1
    k = 1
    tt = [0]
    xt = [1]
    vt = [0]
    VP = [1/2*k*xt[len(xt)-1]*xt[len(xt)-1]]
    VK = [0.5*m*vt[len(vt)-1]*vt[len(vt)-1]]
    V = [VP[len(VP)-1]+VK[len(VK)-1]]
    omega = math.sqrt(k/m)
    stopT = 30 #max Time
    
    while(t <= stopT): 
        tt.append(t)
        xt.append(xt[0]*math.cos(omega*t + fi))
        vt.append(-xt[0]*omega*math.sin(omega*t + fi))
        VP.append(1/2*k*xt[len(xt)-1]*xt[len(xt)-1])
        VK.append(0.5*m*vt[len(vt)-1]*vt[len(vt)-1])
        V.append(VP[len(VP)-1]+VK[len(VK)-1])
        t += dt
    
    return tt, xt, vt, VP, VK, V
```

```python
def simulate_zad3_Euler(dt):
    k = 1      #spring const
    m = 1      #mass
    xtmp = 1   #postion
    vtmp = 0   #velocity
    t = dt      #time
    stopT = 30 #max Time
    betha = 0.5
    
    vt = [0]
    xt = [1]
    tt = [0]
    VP = [1/2*k*xt[len(xt)-1]*xt[len(xt)-1]]
    VK = [0.5*m*vt[len(vt)-1]*vt[len(vt)-1]]
    V = [VP[len(VP)-1]+VK[len(VK)-1]]
    
    while(True):
        vt.append(vtmp - (dt/m)*(betha*vtmp + k*xtmp))
        xt.append(xtmp + dt*vtmp)
        tt.append(t)
        VP.append(1/2*k*xt[len(xt)-1]*xt[len(xt)-1])
        VK.append(0.5*m*vt[len(vt)-1]*vt[len(vt)-1])
        V.append(VP[len(VP)-1]+VK[len(VK)-1])
        xtmp = xt[len(xt)-1]
        vtmp = vt[len(vt)-1]
        t += dt
        if(t >= stopT):
            break
            
    return tt, xt, vt, VP, VK, V
```

```python
def simulate_zad3_Verlet(dt):
    k = 1      #spring const
    m = 1      #mass
    xtmp = 1   #postion
    xtmp2 = 1  #postion
    vtmp = 0   #velocity
    t = dt      #time
    stopT = 30 #max Time
    betha = 0.5
    
    vt = [0]
    xt = [1]
    tt = [0]
    VP = [1/2*k*xt[len(xt)-1]*xt[len(xt)-1]]
    VK = [0.5*m*vt[len(vt)-1]*vt[len(vt)-1]]
    V = [VP[len(VP)-1]+VK[len(VK)-1]]
    
    while(True):  
        vt.append((xtmp-xtmp2)/dt)
        xt.append(-xtmp2 +2*xtmp - dt*dt*(1/m)*(betha*vtmp+k*xtmp))
        tt.append(t)
        VP.append(1/2*k*xt[len(xt)-1]*xt[len(xt)-1])
        VK.append(0.5*m*vt[len(vt)-1]*vt[len(vt)-1])
        V.append(VP[len(VP)-1]+VK[len(VK)-1])
        xtmp2 = xtmp
        xtmp = xt[len(xt)-1]
        vtmp = vt[len(vt)-1]
        t += dt
        if(t >= stopT):
            break
            
    return tt, xt, vt, VP, VK, V
```

```python
def simulate_zad3_exact(dt):
    t = dt
    fi = 0
    fii = 0
    m = 1
    k = 1
    tt = [0]
    xt = [1]
    vt = [0]
    VP = [1/2*k*xt[len(xt)-1]*xt[len(xt)-1]]
    VK = [0.5*m*vt[len(vt)-1]*vt[len(vt)-1]]
    V = [VP[len(VP)-1]+VK[len(VK)-1]]
    omega = math.sqrt(k/m)
    stopT = 30 #max Time
    betha = 0.5
    lamb = betha/(2*m)
    
    while(t <= stopT): 
        tt.append(t)
        xt.append(xt[0]*math.exp((-betha*t)/(2*m))*math.cos(omega*t + fii))
        vt.append(-lamb*xt[0]*math.exp(-lamb*t)*math.cos(omega*t+fi)-omega*xt[0]*math.exp(-lamb*t)*math.sin(omega*t+fi))
        VP.append(1/2*k*xt[len(xt)-1]*xt[len(xt)-1])
        VK.append(0.5*m*vt[len(vt)-1]*vt[len(vt)-1])
        V.append(VP[len(VP)-1]+VK[len(VK)-1])
        t += dt
    
    return tt, xt, vt, VP, VK, V
```
